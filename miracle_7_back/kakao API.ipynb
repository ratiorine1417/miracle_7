```python
import requests

def get_transit_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/transit/routes?lang=0&format=json&appKey={appKey}"
    
    headers = {
        "Content-Type": "application/json"
    }

    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "count": 1
    }

    response = requests.post(url, headers=headers, json=body)

    if response.status_code == 200:
        data = response.json()
        try:
            path = data["metaData"]["plan"]["itineraries"][0]
            totalTime = path["totalTime"]
            totalDistance = path["totalDistance"]

            print("ğŸš‡ ëŒ€ì¤‘êµí†µ ê²½ë¡œ íƒìƒ‰ ê²°ê³¼")
            print(f"ğŸ•’ ì´ ì†Œìš” ì‹œê°„: {totalTime // 60}ë¶„ {totalTime % 60}ì´ˆ")
            print(f"ğŸ“ ì´ ì´ë™ ê±°ë¦¬: {totalDistance} m\n")

            print("ğŸ“ ê²½ë¡œ ìƒì„¸:")
            for idx, leg in enumerate(path["legs"]):
                section_type = leg["mode"]  # WALK, BUS, SUBWAY
                distance = leg["distance"]
                section_time = leg["sectionTime"]

                if section_type == "WALK":
                    print(f"  {idx+1}. ğŸš¶ ë„ë³´ ì´ë™ - ê±°ë¦¬: {distance}m, ì‹œê°„: {section_time}ì´ˆ")
                elif section_type == "BUS":
                    bus_line = leg["route"]  # ë²„ìŠ¤ ë²ˆí˜¸
                    start = leg["start"]["name"]
                    end = leg["end"]["name"]
                    print(f"  {idx+1}. ğŸšŒ ë²„ìŠ¤ {bus_line} â†’ {start} â†’ {end} ({distance}m, {section_time}ì´ˆ)")
                elif section_type == "SUBWAY":
                    subway_line = leg["route"]  # ì§€í•˜ì²  ë…¸ì„ ëª…
                    start = leg["start"]["name"]
                    end = leg["end"]["name"]
                    print(f"  {idx+1}. ğŸš‡ ì§€í•˜ì²  {subway_line} â†’ {start} â†’ {end} ({distance}m, {section_time}ì´ˆ)")
                else:
                    print(f"  {idx+1}. â“ ê¸°íƒ€ ìˆ˜ë‹¨ ({section_type}) - ê±°ë¦¬: {distance}m")

            return totalDistance, totalTime
        except (KeyError, IndexError):
            print("â— ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ì¢Œí‘œë¥¼ ë‹¤ì‹œ í™•ì¸í•´ ì£¼ì„¸ìš”.")
            return None, None
    else:
        print(f"âŒ API ìš”ì²­ ì‹¤íŒ¨: {response.status_code}")
        print(response.text)
        return None, None


# ì—¬ê¸°ì— ì…ë ¥ê°’ë§Œ ë°”ê¾¸ì„¸ìš”
if __name__ == "__main__":
    # âœ… ì¶œë°œì§€ ê²½ë„, ìœ„ë„ (ì˜ˆ: ìœ„ë©ìŠ¤)
    startX = 127.052427705027
    startY = 37.5125058180695

    # âœ… ë„ì°©ì§€ ê²½ë„, ìœ„ë„ (ì˜ˆ: SKì‰´ë”ìŠ¤)
    endX = 127.098937173285
    endY = 37.4044515123315

    # âœ… ë³¸ì¸ì˜ Tmap ë°œê¸‰ ì•±í‚¤
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"

    # ì‹¤í–‰
    get_transit_route(startX, startY, endX, endY, appKey)

```

```python
import requests

def get_walk_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/tmap/routes/pedestrian?version=1&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "ì¶œë°œì§€",
        "endName": "ë„ì°©ì§€"
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            props = response.json()["features"][0]["properties"]
            return ("ë„ë³´", props["totalDistance"], props["totalTime"])
        except:
            return ("ë„ë³´", None, None)
    return ("ë„ë³´", None, None)

def get_car_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/tmap/routes?version=1&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "reqCoordType": "WGS84GEO",
        "resCoordType": "WGS84GEO",
        "startName": "ì¶œë°œì§€",
        "endName": "ë„ì°©ì§€",
        "searchOption": 0,
        "trafficInfo": "Y"
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            props = response.json()["features"][0]["properties"]
            return ("ìë™ì°¨", props["totalDistance"], props["totalTime"])
        except:
            return ("ìë™ì°¨", None, None)
    return ("ìë™ì°¨", None, None)

def get_transit_route(startX, startY, endX, endY, appKey):
    url = f"https://apis.openapi.sk.com/transit/routes?lang=0&format=json&appKey={appKey}"
    headers = {"Content-Type": "application/json"}
    body = {
        "startX": str(startX),
        "startY": str(startY),
        "endX": str(endX),
        "endY": str(endY),
        "count": 1
    }
    response = requests.post(url, headers=headers, json=body)
    if response.status_code == 200:
        try:
            path = response.json()["metaData"]["plan"]["itineraries"][0]
            return ("ëŒ€ì¤‘êµí†µ", path["totalDistance"], path["totalTime"])
        except:
            return ("ëŒ€ì¤‘êµí†µ", None, None)
    return ("ëŒ€ì¤‘êµí†µ", None, None)

def recommend_best_route(startX, startY, endX, endY, appKey):
    results = []
    for func in [get_walk_route, get_car_route, get_transit_route]:
        method, distance, time = func(startX, startY, endX, endY, appKey)
        if time is not None:
            print(f"ğŸ“ {method} - ê±°ë¦¬: {distance}m / ì‹œê°„: {time // 60}ë¶„ {time % 60}ì´ˆ")
            results.append((method, distance, time))
        else:
            print(f"âš ï¸ {method} ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

    if not results:
        print("âŒ ì„¸ ë°©ì‹ ëª¨ë‘ ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")
        return

    best = min(results, key=lambda x: x[2])  # ì†Œìš”ì‹œê°„ ê¸°ì¤€ ìµœì†Œ
    print("\nâœ… ê°€ì¥ ë¹ ë¥¸ ì´ë™ ìˆ˜ë‹¨ ì¶”ì²œ:")
    print(f"ğŸ‘‰ {best[0]} (ì‹œê°„: {best[2] // 60}ë¶„ {best[2] % 60}ì´ˆ, ê±°ë¦¬: {best[1]}m)")



# ì—¬ê¸°ì— ì…ë ¥ê°’ë§Œ ë°”ê¾¸ì„¸ìš”
if __name__ == "__main__":
    # âœ… ì¶œë°œì§€ ê²½ë„, ìœ„ë„ (ì˜ˆ: ìœ„ë©ìŠ¤)
    startX = 127.052427705027
    startY = 37.5125058180695

    # âœ… ë„ì°©ì§€ ê²½ë„, ìœ„ë„ (ì˜ˆ: SKì‰´ë”ìŠ¤)
    endX = 127.098937173285
    endY = 37.4044515123315

    # âœ… ë³¸ì¸ì˜ Tmap ë°œê¸‰ ì•±í‚¤
    appKey = "KXgnElYbnf9P4rPyZzanN91cHMyScabF1ZY4ifZR"

    recommend_best_route(startX, startY, endX, endY, appKey)
```
